```python
import seaborn as sns #this is the plotting library I'll be using 
import pandas as pd #"as pd" means that we can use the abbreviation in commands
import matplotlib.pyplot as plt #we need Matplotlib for setting the labels in the Seaborn graphs
%matplotlib inline

df_srv = pd.read_csv('survey.csv', sep=',')
df_srv.head()
```

```python
df_stp = pd.read_csv('steps.csv', sep=';')
df_stp.head()
```

```python
df_mrg = pd.merge(df_srv,df_stp , on = 'id')
df_mrg.head()
```

```python
df_mrg.info() #Get information on the variables
```

```python
sns.distplot(df_mrg["height"], kde=False) #kde=False means we don't get an overlying density plot
plt.title("Height distribution of participants")
plt.xlabel("Height")
plt.ylabel("Count")
plt.show()
```

```python
sns.set_style('whitegrid') 
sns.kdeplot(df_mrg['height'])
```

```python
sns.distplot(df_mrg["weight"], kde=False)
plt.title("Weight distribution of participants")
plt.xlabel("Weight")
plt.ylabel("Count")
plt.show()
```

```python
sns.set_style('whitegrid') 
sns.kdeplot(df_mrg['weight'])
```

```python
# calculate values statistics for height
mean = str(float(df_mrg[['height']].mean()))
median = str(float(df_mrg[['height']].median()))
mode = str(df_mrg[['height']].mode()['height'][0])
variance = str(float(df_mrg[['height']].var()))
stdev = str(float(df_mrg[['height']].std()))
```

```python
print('median: ', median)
print('mode: ' , mode)
print('mean: ' , mean)
print('standard deviation: ', stdev)
print('variance: ', variance)
```

```python
# calculate values for weight
mean = str(float(df_mrg[['weight']].mean()))
median = str(float(df_mrg[['weight']].median()))
mode = str(df_mrg[['weight']].mode()['weight'][0])
variance = str(float(df_mrg[['weight']].var()))
stdev = str(float(df_mrg[['weight']].std()))
```

```python
print('median: ', median)
print('mode: ' , mode)
print('mean: ' , mean)
print('standard deviation: ', stdev)
print('variance: ', variance)
```

```python
mean_steps = df_mrg.loc[:,"20-6-2013":"13-5-2014"].mean() #get the mean over the columns (all the dates) and store
mean_steps.index #the indexes now contain the dates
```

```python
plt.figure(figsize=(25, 6)) #set the figure ("canvas") size
plot = sns.lineplot(x=mean_steps.index, y=mean_steps) #create the plot

for ind, label in enumerate(plot.get_xticklabels()):
    if ind % 20 == 0:  # every 10th label is kept
        label.set_visible(True)
    else:
        label.set_visible(False)

plt.title("Mean steps per date")
plt.xlabel("Date")
plt.ylabel("steps")

plt.show()
```
